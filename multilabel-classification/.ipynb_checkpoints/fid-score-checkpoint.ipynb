```python
# Resize all images

from PIL import Image
import os, sys

```

```python
def resize_images_in_dir_to_dir(dir_name_source, dir_name_dest):
    n = 0
    files = os.listdir(dir_name_source)
    for file_name in files:
        if file_name.endswith('.jpg'):
            source_file = dir_name_source + file_name
            dest_file  = dir_name_dest + file_name
            im = Image.open(source_file)
            imResize = im.resize((512,512), Image.ANTIALIAS)
            imResize.save(dest_file, 'JPEG', quality=95, subsampling=0)
            n=n+1
            print("Processed [", n,"] files.")

```

```python
test_path = "/scratch/rm5707/roco-dataset-master/data/test/radiology/images/"
new_test_path = "/scratch/rm5707/roco-dataset-master/data/test/radiology/images-resized/"

resize_images_in_dir_to_dir(test_path, new_test_path)
```

```python

test_gen_path = "/scratch/rm5707/test-generated/test/"
new_test_gen_path = "/scratch/rm5707/test-generated/test-resized/"

resize_images_in_dir_to_dir(test_gen_path, new_test_gen_path)
```
