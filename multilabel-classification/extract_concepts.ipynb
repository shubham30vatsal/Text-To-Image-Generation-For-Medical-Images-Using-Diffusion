```python
# only radiology
roco_captions_file_train = "/scratch/rm5707/roco-dataset-master/data/train/radiology/cuis.txt"
roco_captions_file_val   = "/scratch/rm5707/roco-dataset-master/data/validation/radiology/cuis.txt"
roco_captions_file_test  = "/scratch/rm5707/roco-dataset-master/data/test/radiology/cuis.txt"
```

```python
file1 = open(roco_captions_file_train, 'r')
count = 0
cuids_dict = dict()
max_cuids_for_image = 0
sum_cuids = 0.

while True:
    count += 1
    

    # Get next line from file
    line = file1.readline()
    
    if not line:
        break
    line_splits = line.split('\t')

    image_id = line_splits[0]
    if(line_splits[1] == ''):
        cuids = line_splits[2:-1]
    else :
        cuids = line_splits[1:-1]

    if(count == 1):
        print(image_id)
        print(cuids)
        
    sum_cuids = sum_cuids + len(cuids)
    if(len(cuids) > max_cuids_for_image):
        max_cuids_for_image = len(cuids)
        
    
    for cuid in cuids:
        cuids_dict[cuid] = cuids_dict.get(cuid, 0) + 1
        
    # create cuids file for per image information

    # if line is empty
    # end of file is reached

    #print("Line{}: {}".format(count, line.strip()))
  
file1.close()
print(count)
```

```python
sorted_cuids_by_count = {k: v for k, v in sorted(cuids_dict.items(), key=lambda item: item[1], reverse=True)}
sorted_cuids_by_count
```

```python
import itertools 
N = 100
top_N_cuids = dict(itertools.islice(sorted_cuids_by_count.items(), N)) 
top_N_cuids
```

```python
top_N_cuids_list = list(top_N_cuids.keys())
#'C1837463' in top_N_cuids_list
```

```python
concepts_file_name = "concepts_top_"+ str(N) +".txt"
with open(concepts_file_name, 'w') as f:
    f.write('concept\n')
    for cui in top_N_cuids:
        f.write(cui + "\n")
```

```python
file1 = open(roco_captions_file_val, 'r')
count = 0
image_to_cuids_dict = dict()
max_cuids_per_image = 0

while True:
    count += 1
    

    # Get next line from file
    line = file1.readline()
    
    if not line:
        break

    line_splits = line.split('\t')

    image_id = line_splits[0]
    if(line_splits[1] == ''):
        cuids = line_splits[2:-1]
    else :
        cuids = line_splits[1:-1]
        
    cuids_filtered_on_top_list = []
    
    for cuid in cuids:
        if cuid in top_N_cuids_list:
            cuids_filtered_on_top_list.append(cuid)
        
    if(max_cuids_per_image < len(cuids_filtered_on_top_list)):
        max_cuids_per_image = len(cuids_filtered_on_top_list)
    #print("Line{}: {}".format(count, line.strip()))
    if(len(cuids_filtered_on_top_list) > 0):
        image_path = '/scratch/rm5707/roco-dataset-master/data/validation/radiology/images/' + image_id + '.jpg'
        if(os.path.exists(image_path)):
            image_to_cuids_dict[image_id] = cuids_filtered_on_top_list
        else:
            print("Skipping : ", image_path)
  
file1.close()
print(count)
```

```python
cuis_file_name = "cuis_top_val_"+ str(N) +".txt"
with open(cuis_file_name, 'w') as f:
    f.write('ID,cuis\n')
    for image_id, cuis in image_to_cuids_dict.items():
        cuis_col = ";".join(cuis)
        f.write(",".join([image_id, cuis_col]) + "\n")
```
